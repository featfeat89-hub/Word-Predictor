```python
import urllib.request

urls = [
    # Previous books
    "https://www.gutenberg.org/files/1342/1342-0.txt",
    "https://www.gutenberg.org/files/84/84-0.txt",
    "https://www.gutenberg.org/files/11/11-0.txt",
    "https://www.gutenberg.org/files/1661/1661-0.txt",
    "https://www.gutenberg.org/files/98/98-0.txt",
    "https://www.gutenberg.org/files/2701/2701-0.txt",
    "https://www.gutenberg.org/files/174/174-0.txt",
    "https://www.gutenberg.org/files/76/76-0.txt",
    "https://www.gutenberg.org/files/1400/1400-0.txt",
    "https://www.gutenberg.org/files/2600/2600-0.txt",
    "https://www.gutenberg.org/files/345/345-0.txt",
    "https://www.gutenberg.org/files/1260/1260-0.txt",
    "https://www.gutenberg.org/files/768/768-0.txt",
    "https://www.gutenberg.org/files/730/730-0.txt",
    "https://www.gutenberg.org/files/2591/2591-0.txt",
    "https://www.gutenberg.org/files/1727/1727-0.txt",
    "https://www.gutenberg.org/files/6130/6130-0.txt",
    "https://www.gutenberg.org/files/2554/2554-0.txt",
    "https://www.gutenberg.org/files/28054/28054-0.txt",
    "https://www.gutenberg.org/files/36/36-0.txt",
    "https://www.gutenberg.org/files/35/35-0.txt",
    "https://www.gutenberg.org/files/5200/5200-0.txt",
    "https://www.gutenberg.org/files/4300/4300-0.txt",
    "https://www.gutenberg.org/files/1399/1399-0.txt",
    "https://www.gutenberg.org/files/219/219-0.txt",
    "https://www.gutenberg.org/files/135/135-0.txt",
    "https://www.gutenberg.org/files/2500/2500-0.txt",
    "https://www.gutenberg.org/files/514/514-0.txt",
    "https://www.gutenberg.org/files/158/158-0.txt",
    "https://www.gutenberg.org/files/161/161-0.txt",
    "https://www.gutenberg.org/files/105/105-0.txt",
    "https://www.gutenberg.org/files/766/766-0.txt",
    "https://www.gutenberg.org/files/786/786-0.txt",
    "https://www.gutenberg.org/files/110/110-0.txt",
    "https://www.gutenberg.org/files/209/209-0.txt",
    "https://www.gutenberg.org/files/160/160-0.txt",
    "https://www.gutenberg.org/files/233/233-0.txt",
    "https://www.gutenberg.org/files/215/215-0.txt",
    "https://www.gutenberg.org/files/43/43-0.txt",
    "https://www.gutenberg.org/files/120/120-0.txt",
    "https://www.gutenberg.org/files/244/244-0.txt",
    "https://www.gutenberg.org/files/2097/2097-0.txt",
    "https://www.gutenberg.org/files/1155/1155-0.txt",
    "https://www.gutenberg.org/files/3913/3913-0.txt",
    "https://www.gutenberg.org/files/45/45-0.txt",
    "https://www.gutenberg.org/files/514/514-0.txt",
    "https://www.gutenberg.org/files/1513/1513-0.txt",
    "https://www.gutenberg.org/files/1524/1524-0.txt",
    "https://www.gutenberg.org/files/1532/1532-0.txt",
    "https://www.gutenberg.org/files/1533/1533-0.txt",
    "https://www.gutenberg.org/files/816/816-0.txt",

    # hundreds more below
    "https://www.gutenberg.org/files/2814/2814-0.txt",   # Dubliners
    "https://www.gutenberg.org/files/4363/4363-0.txt",   # Beyond Good and Evil
    "https://www.gutenberg.org/files/1998/1998-0.txt",   # Thus Spoke Zarathustra
    "https://www.gutenberg.org/files/3207/3207-0.txt",   # Leviathan
    "https://www.gutenberg.org/files/1322/1322-0.txt",   # Wealth of Nations
    "https://www.gutenberg.org/files/3300/3300-0.txt",   # Federalist Papers
    "https://www.gutenberg.org/files/2412/2412-0.txt",   # Art of War
    "https://www.gutenberg.org/files/103/103-0.txt",     # Around the World in 80 Days
    "https://www.gutenberg.org/files/164/164-0.txt",     # Twenty Thousand Leagues
    "https://www.gutenberg.org/files/367/367-0.txt",     # The Mysterious Island
    "https://www.gutenberg.org/files/1184/1184-0.txt",   # Count of Monte Cristo full
    "https://www.gutenberg.org/files/1259/1259-0.txt",   # Twenty Years After
    "https://www.gutenberg.org/files/3825/3825-0.txt",   # My Antonia
    "https://www.gutenberg.org/files/1321/1321-0.txt",   # O Pioneers
    "https://www.gutenberg.org/files/19942/19942-0.txt", # Pollyanna
    "https://www.gutenberg.org/files/47/47-0.txt",       # Anne of Avonlea
    "https://www.gutenberg.org/files/51/51-0.txt",       # Anne of the Island
    "https://www.gutenberg.org/files/600/600-0.txt",     # Notes from Underground
    "https://www.gutenberg.org/files/2413/2413-0.txt",   # Dead Souls
    "https://www.gutenberg.org/files/580/580-0.txt",     # Pickwick Papers
    "https://www.gutenberg.org/files/153/153-0.txt",     # Mayor of Casterbridge
    "https://www.gutenberg.org/files/4276/4276-0.txt",   # Return of the Native
    "https://www.gutenberg.org/files/834/834-0.txt",     # Memoirs of Sherlock Holmes
    "https://www.gutenberg.org/files/108/108-0.txt",     # Return of Sherlock Holmes
    "https://www.gutenberg.org/files/2852/2852-0.txt",   # Casebook of Sherlock Holmes
    "https://www.gutenberg.org/files/863/863-0.txt",     # Edwin Drood

    # Brand new additions - pushing toward 50 million tokens
    "https://www.gutenberg.org/files/2680/2680-0.txt",   # Meditations - Marcus Aurelius
    "https://www.gutenberg.org/files/1497/1497-0.txt",   # The Republic - Plato full
    "https://www.gutenberg.org/files/1642/1642-0.txt",   # On War - Clausewitz
    "https://www.gutenberg.org/files/3600/3600-0.txt",   # Autobiography of Ben Franklin
    "https://www.gutenberg.org/files/1232/1232-0.txt",   # The Prince - Machiavelli
    "https://www.gutenberg.org/files/1404/1404-0.txt",   # Magna Carta and documents
    "https://www.gutenberg.org/files/2009/2009-0.txt",   # The Communist Manifesto
    "https://www.gutenberg.org/files/61/61-0.txt",       # The Communist Manifesto alt
    "https://www.gutenberg.org/files/5827/5827-0.txt",   # The Souls of Black Folk
    "https://www.gutenberg.org/files/23/23-0.txt",       # Narrative of Frederick Douglass
    "https://www.gutenberg.org/files/131/131-0.txt",     # The Sun Also Rises (alt)
    "https://www.gutenberg.org/files/526/526-0.txt",     # The Idiot - Dostoevsky
    "https://www.gutenberg.org/files/2197/2197-0.txt",   # The Possessed - Dostoevsky
    "https://www.gutenberg.org/files/4986/4986-0.txt",   # The Gambler - Dostoevsky
    "https://www.gutenberg.org/files/41866/41866-0.txt", # The Eternal Husband
    "https://www.gutenberg.org/files/1399/1399-0.txt",   # Anna Karenina alt
    "https://www.gutenberg.org/files/243/243-0.txt",     # The Jungle Book
    "https://www.gutenberg.org/files/236/236-0.txt",     # The Second Jungle Book
    "https://www.gutenberg.org/files/2610/2610-0.txt",   # Kim - Kipling
    "https://www.gutenberg.org/files/1952/1952-0.txt",   # The Yellow Wallpaper
    "https://www.gutenberg.org/files/160/160-0.txt",     # The Scarlet Letter alt
    "https://www.gutenberg.org/files/33/33-0.txt",       # The Scarlet Letter full
    "https://www.gutenberg.org/files/77/77-0.txt",       # The Man Who Was Thursday
    "https://www.gutenberg.org/files/1064/1064-0.txt",   # The Napoleon of Notting Hill
    "https://www.gutenberg.org/files/87/87-0.txt",       # Father Brown stories
    "https://www.gutenberg.org/files/223/223-0.txt",     # The Invisible Man - Wells
    "https://www.gutenberg.org/files/159/159-0.txt",     # The Island of Doctor Moreau
    "https://www.gutenberg.org/files/718/718-0.txt",     # The First Men in the Moon
    "https://www.gutenberg.org/files/1743/1743-0.txt",   # Tono Bungay - Wells
    "https://www.gutenberg.org/files/3362/3362-0.txt",   # The History of Mr Polly
    "https://www.gutenberg.org/files/4217/4217-0.txt",   # Ann Veronica - Wells
    "https://www.gutenberg.org/files/1268/1268-0.txt",   # Anticipations - Wells
    "https://www.gutenberg.org/files/622/622-0.txt",     # The Food of the Gods - Wells
    "https://www.gutenberg.org/files/7083/7083-0.txt",   # Love and Mr Lewisham
    "https://www.gutenberg.org/files/2295/2295-0.txt",   # The Sea Lady - Wells
    "https://www.gutenberg.org/files/1046/1046-0.txt",   # Kipps - Wells
    "https://www.gutenberg.org/files/3870/3870-0.txt",   # In the Days of the Comet
    "https://www.gutenberg.org/files/9830/9830-0.txt",   # Bealby - Wells
    "https://www.gutenberg.org/files/1708/1708-0.txt",   # Twelve Stories and a Dream
    "https://www.gutenberg.org/files/456/456-0.txt",     # The Stolen Bacillus - Wells
    "https://www.gutenberg.org/files/30383/30383-0.txt", # The Door in the Wall - Wells
    "https://www.gutenberg.org/files/11/11-0.txt",       # Alice alt
    "https://www.gutenberg.org/files/12/12-0.txt",       # Through the Looking Glass
    "https://www.gutenberg.org/files/620/620-0.txt",     # The Hunting of the Snark
    "https://www.gutenberg.org/files/1064/1064-0.txt",   # Sylvie and Bruno
    "https://www.gutenberg.org/files/4763/4763-0.txt",   # Phantasmagoria
    "https://www.gutenberg.org/files/651/651-0.txt",     # A Tangled Tale
    "https://www.gutenberg.org/files/13/13-0.txt",       # Paradise Lost
    "https://www.gutenberg.org/files/20/20-0.txt",       # The Narrative of Arthur Gordon Pym
    "https://www.gutenberg.org/files/2148/2148-0.txt",   # Eureka - Poe
    "https://www.gutenberg.org/files/932/932-0.txt",     # The Fall of the House of Usher
    "https://www.gutenberg.org/files/2147/2147-0.txt",   # Complete Poetical Works of Poe
    "https://www.gutenberg.org/files/1064/1064-0.txt",   # Complete Works alt
    "https://www.gutenberg.org/files/408/408-0.txt",     # The Autocrat of the Breakfast Table
    "https://www.gutenberg.org/files/2800/2800-0.txt",   # Uncle Tom's Cabin
    "https://www.gutenberg.org/files/6688/6688-0.txt",   # The Souls of Black Folk alt
    "https://www.gutenberg.org/files/829/829-0.txt",     # Gulliver's Travels
    "https://www.gutenberg.org/files/1080/1080-0.txt",   # A Modest Proposal alt
    "https://www.gutenberg.org/files/17157/17157-0.txt", # The Journal to Stella
    "https://www.gutenberg.org/files/623/623-0.txt",     # Don Quixote
    "https://www.gutenberg.org/files/996/996-0.txt",     # Don Quixote full
    "https://www.gutenberg.org/files/2000/2000-0.txt",   # Don Quixote alt
    "https://www.gutenberg.org/files/5851/5851-0.txt",   # The Arabian Nights Vol 1
    "https://www.gutenberg.org/files/7243/7243-0.txt",   # The Arabian Nights Vol 2
    "https://www.gutenberg.org/files/8655/8655-0.txt",   # The Arabian Nights Vol 3
    "https://www.gutenberg.org/files/9206/9206-0.txt",   # The Arabian Nights Vol 4
    "https://www.gutenberg.org/files/128/128-0.txt",     # The Red Badge of Courage
    "https://www.gutenberg.org/files/174/174-0.txt",     # Dorian Gray alt
    "https://www.gutenberg.org/files/4085/4085-0.txt",   # De Profundis - Wilde
    "https://www.gutenberg.org/files/790/790-0.txt",     # A House of Pomegranates
    "https://www.gutenberg.org/files/773/773-0.txt",     # The Picture of Dorian Gray alt
    "https://www.gutenberg.org/files/921/921-0.txt",     # The Happy Prince - Wilde
    "https://www.gutenberg.org/files/1017/1017-0.txt",   # An Ideal Husband - Wilde
    "https://www.gutenberg.org/files/790/790-0.txt",     # Lady Windermere's Fan
    "https://www.gutenberg.org/files/1309/1309-0.txt",   # A Woman of No Importance
    "https://www.gutenberg.org/files/902/902-0.txt",     # Salome - Wilde
    "https://www.gutenberg.org/files/4297/4297-0.txt",   # The Ballad of Reading Gaol
    "https://www.gutenberg.org/files/301/301-0.txt",     # The Canterbury Tales
    "https://www.gutenberg.org/files/22/22-0.txt",       # The Pilgrim's Progress
    "https://www.gutenberg.org/files/3207/3207-0.txt",   # Leviathan alt
    "https://www.gutenberg.org/files/10/10-0.txt",       # The Bible - King James
    "https://www.gutenberg.org/files/8800/8800-0.txt",   # The Koran
    "https://www.gutenberg.org/files/7370/7370-0.txt",   # Democracy in America
    "https://www.gutenberg.org/files/815/815-0.txt",     # Walden - Thoreau
    "https://www.gutenberg.org/files/16/16-0.txt",       # Peter Pan
    "https://www.gutenberg.org/files/19033/19033-0.txt", # Peter and Wendy
    "https://www.gutenberg.org/files/5765/5765-0.txt",   # The Wind in the Willows
    "https://www.gutenberg.org/files/23172/23172-0.txt", # Watership Down alt
    "https://www.gutenberg.org/files/74/74-0.txt",       # Tom Sawyer alt
    "https://www.gutenberg.org/files/86/86-0.txt",       # A Connecticut Yankee
    "https://www.gutenberg.org/files/245/245-0.txt",     # Roughing It - Twain
    "https://www.gutenberg.org/files/3177/3177-0.txt",   # Life on the Mississippi
    "https://www.gutenberg.org/files/119/119-0.txt",     # The Innocents Abroad
    "https://www.gutenberg.org/files/102/102-0.txt",     # The Prince and the Pauper
    "https://www.gutenberg.org/files/3176/3176-0.txt",   # Personal Recollections of Joan of Arc
    "https://www.gutenberg.org/files/70/70-0.txt",       # The Tragedy of Pudd'nhead Wilson
    "https://www.gutenberg.org/files/1837/1837-0.txt",   # The Mysterious Stranger
    "https://www.gutenberg.org/files/142/142-0.txt",     # The $30,000 Bequest
    "https://www.gutenberg.org/files/3251/3251-0.txt",   # What Is Man - Twain
    "https://www.gutenberg.org/files/2895/2895-0.txt",   # Is Shakespeare Dead - Twain
    "https://www.gutenberg.org/files/325/325-0.txt",     # Following the Equator
    "https://www.gutenberg.org/files/1289/1289-0.txt",   # The Gilded Age - Twain
    "https://www.gutenberg.org/files/3182/3182-0.txt",   # A Double Barrelled Detective
    "https://www.gutenberg.org/files/91/91-0.txt",       # Tom Sawyer Abroad
    "https://www.gutenberg.org/files/93/93-0.txt",       # Tom Sawyer Detective
    "https://www.gutenberg.org/files/1044/1044-0.txt",   # Extracts from Adam's Diary
    "https://www.gutenberg.org/files/8525/8525-0.txt",   # Eve's Diary
    "https://www.gutenberg.org/files/572/572-0.txt",     # Captain Stormfield's Visit to Heaven
    "https://www.gutenberg.org/files/3188/3188-0.txt",   # Sketches New and Old - Twain
    "https://www.gutenberg.org/files/2895/2895-0.txt",   # Christian Science - Twain
    "https://www.gutenberg.org/files/2786/2786-0.txt",   # The Man That Corrupted Hadleyburg
    "https://www.gutenberg.org/files/7100/7100-0.txt",   # Complete Essays - Montaigne
    "https://www.gutenberg.org/files/3600/3600-0.txt",   # Ben Franklin Autobiography alt
    "https://www.gutenberg.org/files/20203/20203-0.txt", # The Education of Henry Adams
    "https://www.gutenberg.org/files/2086/2086-0.txt",   # Up From Slavery - Booker T Washington
    "https://www.gutenberg.org/files/17784/17784-0.txt", # My Bondage and My Freedom - Douglass
    "https://www.gutenberg.org/files/1168/1168-0.txt",   # The Life of Olaudah Equiano
    "https://www.gutenberg.org/files/6688/6688-0.txt",   # The Souls of Black Folk - Du Bois
    "https://www.gutenberg.org/files/408/408-0.txt",     # Essays - Emerson
    "https://www.gutenberg.org/files/16643/16643-0.txt", # Nature - Emerson
    "https://www.gutenberg.org/files/2944/2944-0.txt",   # Self Reliance - Emerson
    "https://www.gutenberg.org/files/15/15-0.txt",       # Moby Dick alt
    "https://www.gutenberg.org/files/2489/2489-0.txt",   # Bartleby the Scrivener - Melville
    "https://www.gutenberg.org/files/10712/10712-0.txt", # Billy Budd - Melville
    "https://www.gutenberg.org/files/1900/1900-0.txt",   # The Awakening alt - Chopin
    "https://www.gutenberg.org/files/160/160-0.txt",     # The House of Mirth - Wharton
    "https://www.gutenberg.org/files/284/284-0.txt",     # The Age of Innocence - Wharton
    "https://www.gutenberg.org/files/312/312-0.txt",     # Ethan Frome - Wharton
    "https://www.gutenberg.org/files/174/174-0.txt",     # The Custom of the Country
    "https://www.gutenberg.org/files/541/541-0.txt",     # The Valley of Decision - Wharton
    "https://www.gutenberg.org/files/1237/1237-0.txt",   # Daisy Miller - Henry James
    "https://www.gutenberg.org/files/209/209-0.txt",     # The Turn of the Screw - Henry James
    "https://www.gutenberg.org/files/177/177-0.txt",     # Washington Square - Henry James
    "https://www.gutenberg.org/files/432/432-0.txt",     # The Bostonians - Henry James
    "https://www.gutenberg.org/files/2850/2850-0.txt",   # The American - Henry James
    "https://www.gutenberg.org/files/1202/1202-0.txt",   # The Europeans - Henry James
    "https://www.gutenberg.org/files/2388/2388-0.txt",   # Roderick Hudson - Henry James
    "https://www.gutenberg.org/files/7025/7025-0.txt",   # The Portrait of a Lady - Henry James
    "https://www.gutenberg.org/files/1183/1183-0.txt",   # The Wings of the Dove - James
    "https://www.gutenberg.org/files/1179/1179-0.txt",   # The Ambassadors - Henry James
    "https://www.gutenberg.org/files/1280/1280-0.txt",   # The Golden Bowl - Henry James
    "https://www.gutenberg.org/files/268/268-0.txt",     # Sister Carrie - Dreiser
    "https://www.gutenberg.org/files/1164/1164-0.txt",   # Jennie Gerhardt - Dreiser
    "https://www.gutenberg.org/files/1155/1155-0.txt",   # The Financier - Dreiser
    "https://www.gutenberg.org/files/370/370-0.txt",     # Main Street - Sinclair Lewis
    "https://www.gutenberg.org/files/1156/1156-0.txt",   # Babbitt - Sinclair Lewis
    "https://www.gutenberg.org/files/3078/3078-0.txt",   # Arrowsmith - Sinclair Lewis
    "https://www.gutenberg.org/files/2003/2003-0.txt",   # O Henry Short Stories
    "https://www.gutenberg.org/files/1467/1467-0.txt",   # Cabbages and Kings - O Henry
    "https://www.gutenberg.org/files/1583/1583-0.txt",   # The Four Million - O Henry
    "https://www.gutenberg.org/files/2776/2776-0.txt",   # Heart of the West - O Henry
    "https://www.gutenberg.org/files/1800/1800-0.txt",   # The Voice of the City - O Henry
    "https://www.gutenberg.org/files/1594/1594-0.txt",   # The Trimmed Lamp - O Henry
    "https://www.gutenberg.org/files/1870/1870-0.txt",   # Strictly Business - O Henry
    "https://www.gutenberg.org/files/2536/2536-0.txt",   # Whirligigs - O Henry
    "https://www.gutenberg.org/files/1643/1643-0.txt",   # Roads of Destiny - O Henry
    "https://www.gutenberg.org/files/1601/1601-0.txt",   # Options - O Henry
    "https://www.gutenberg.org/files/2777/2777-0.txt",   # Sixes and Sevens - O Henry
    "https://www.gutenberg.org/files/1666/1666-0.txt",   # Rolling Stones - O Henry
    "https://www.gutenberg.org/files/2178/2178-0.txt",   # Waifs and Strays - O Henry
    "https://www.gutenberg.org/files/7256/7256-0.txt",   # Postscripts - O Henry
    "https://www.gutenberg.org/files/1688/1688-0.txt",   # Aftermath - O Henry
    "https://www.gutenberg.org/files/9659/9659-0.txt",   # Collected Poems - Whitman
    "https://www.gutenberg.org/files/1322/1322-0.txt",   # Leaves of Grass - Whitman
    "https://www.gutenberg.org/files/1065/1065-0.txt",   # Song of Myself - Whitman
    "https://www.gutenberg.org/files/18886/18886-0.txt", # Complete Poems - Emily Dickinson
    "https://www.gutenberg.org/files/12242/12242-0.txt", # Poems - Emily Dickinson
    "https://www.gutenberg.org/files/2600/2600-0.txt",   # War and Peace again - it is enormous
    "https://www.gutenberg.org/files/4300/4300-0.txt",   # Ulysses again - it is enormous
    "https://www.gutenberg.org/files/1184/1184-0.txt",   # Monte Cristo again - it is enormous
    "https://www.gutenberg.org/files/28054/28054-0.txt", # Brothers Karamazov again
    "https://www.gutenberg.org/files/135/135-0.txt",     # Les Miserables again
]

MY_TEXT = ""
seen = set()
for url in urls:
    if url in seen:
        continue
    seen.add(url)
    try:
        title = url.split("/")[-2]
        print(f"Downloading {title}...")
        with urllib.request.urlopen(url) as response:
            MY_TEXT += response.read().decode("utf-8", errors="ignore")
        print(f"  ✅ {len(MY_TEXT):,} chars so far")
    except Exception as e:
        print(f"  ❌ Failed: {e}")

print(f"\n📚 Total characters: {len(MY_TEXT):,}")
print(f"📝 Estimated words: {len(MY_TEXT.split()):,}")
```

```python
def train(text, n=2):
    tokens = nltk.word_tokenize(text.lower())
    print(f"🔢 Total tokens to process: {len(tokens):,}")
    print("⏳ Training... this will take a few minutes!")
    model = defaultdict(list)
    for i in range(len(tokens) - n):
        key = tuple(tokens[i:i+n])
        next_word = tokens[i+n]
        model[key].append(next_word)
    return model, n

model, N = train(MY_TEXT, n=2)
print(f"\n✅ Trained! Learned {len(model):,} unique word patterns.")
```

```python
def generate_sentence(start_words, length=15):
    tokens = nltk.word_tokenize(start_words.lower())
    for _ in range(length):
        key = tuple(tokens[-N:])
        options = model.get(key, [])
        if not options:
            break
        tokens.append(random.choice(options))
    return " ".join(tokens)

print(generate_sentence("the sun"))
print(generate_sentence("the day"))
print(generate_sentence("a gentle"))
```
